```python
import matplotlib.pyplot as plt

import pandas as pd

df = pd.read_csv("C:/Users/user/.kaggle/KaggleDataSets/PlayStation Sales and Metadata (PS3PS4PS5) (Oct 2025).csv")

Na_sales =df["NA Sales"]
Jap_sales = df["Japan Sales"]

plt.figure(figsize = (10,10))
plt.plot(Na_sales,Jap_sales)
plt.xticks(rotation = 90)
plt.xlabel("NA Sales")
plt.ylabel("Japan Sales")
plt.title("Japan Sales vs. NA Sales over Time")
plt.show()
```
